#### Write a Python program to create a plot (lines) of total deaths, confirmed, recovered and active cases Country wise where deaths greater than 150.

```python
# df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
filtered = df[df['Deaths'] > 150]
p = filtered.groupby(df['Country/Region']).sum(numeric_only=True).reset_index()
p

plt.figure(figsize=(15,10))
plt.plot(p['Country/Region'] ,p['Deaths'] , label='Deaths' , marker='o')
plt.plot(p['Country/Region'] ,p['Confirmed'] , label='Confirmed', marker='o')
plt.plot(p['Country/Region'] ,p['Recovered'] , label='Recovered' , marker='o')
plt.plot(p['Country/Region'] ,p['Active'] , label='Active' , marker='o')
plt.xlabel('Country/Region')
plt.ylabel('Num of Cases')
plt.title('COVID-19 Cases in Country where daths > 150')

plt.legend()
plt.grid(True)

plt.show()
```
